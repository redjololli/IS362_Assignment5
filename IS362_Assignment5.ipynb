### Question 3 
- On February 12th, 2013, which New York area airport had the windiest weather?

```python
# import the data for weather
import pandas as pd
weather = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
```

```python

weather[(weather['year'] == 2013)&(weather['month'] == 2)& (weather['day'] == 12)&
        (weather.wind_gust < 500)].sort_values(by = ['wind_gust'], ascending=False).head(5)
```


- Ans: Based on the findings Newark Airport had the windest weather; but which New York area airport had the windiest weather? LGA airport located in NYC was the windiest airport on February 12th, 2013.
